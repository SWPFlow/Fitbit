```python
%matplotlib inline
from fitbittools.fbdata import FitBitData                                              
from fitbittools.plotters import Plotter
```

```python
filename="/Users/frenchwr/Computing/Python/Fitbit/data/fitbit_export_20140710.csv"
fb_obj = FitBitData(filename)
activity = fb_obj.activity.pd_data
sleep = fb_obj.sleep.pd_data
plotter = Plotter()
```

```python
df = activity[['Minutes Sedentary','Minutes Lightly Active']]
print df.head()
plotter.plot_data_frame(df,kind='bar',stacked=True,figsize="Large")
```

```python
plotter.plot_data_frame(activity,figsize="Large")
```

```python
plotter.plot_data_frame(sleep,kind='bar',stacked=True,figsize="Large")
```

```python
plotter.plot_data_frame(activity,kind='area',figsize="Large")
```

```python
plotter.plot_data_frame(activity,x='Minutes Sedentary',y='Minutes Very Active',kind='hexbin',figsize="Large")
```

```python
plotter.plot_data_frame(sleep,kind='hist',stacked=True,figsize="Large")
```

```python
plotter.plot_data_frame(activity,kind='box',figsize="Large")
```

```python
plotter.plot_data_frame(activity,y=['Minutes Very Active','Minutes Lightly Active'],kind='pie',figsize="Large")
```
